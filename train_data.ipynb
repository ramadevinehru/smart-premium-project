```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "train.csv" 
df = pd.read_csv(file_path)

print("Dataset Info:")
print(df.info())

missing_values = df.isnull().sum().sort_values(ascending=False)
print("\nMissing Values:")
print(missing_values[missing_values > 0])

# Plot missing values
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_values.index, y=missing_values.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.ylabel("Count of Missing Values")
plt.title("Missing Values in Each Column")
plt.show()

print("\nSummary Statistics:")
print(df.describe())

num_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col].dropna(), bins=30, kde=True, color="blue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df[num_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Load dataset
df = pd.read_csv("train.csv")  

df['Policy Start Year'] = pd.to_datetime(df['Policy Start Date'], errors='coerce').dt.year
df.drop(['Policy Start Date'], axis=1, inplace=True)

df.fillna({
    'Number of Dependents': df['Number of Dependents'].median(),
    'Credit Score': df['Credit Score'].median(),
    'Health Score': df['Health Score'].median(),
    'Previous Claims': df['Previous Claims'].median(),
    'Vehicle Age': df['Vehicle Age'].median()
}, inplace=True)

df.drop(['Customer Feedback'], axis=1, inplace=True)

# One-hot encoding for categorical variables
df = pd.get_dummies(df, columns=[
    'Gender', 'Marital Status', 'Education Level', 'Occupation', 'Location', 
    'Policy Type', 'Smoking Status', 'Exercise Frequency', 'Property Type'
], dtype=int)

# Handle any remaining NaN values after encoding
df.fillna(0, inplace=True) 

# Feature Scaling (Standardization) for numerical variables
df['Annual Income'] = (df['Annual Income'] - df['Annual Income'].mean()) / df['Annual Income'].std()
df['Premium Amount'] = (df['Premium Amount'] - df['Premium Amount'].mean()) / df['Premium Amount'].std()


X = df.drop(['Premium Amount'], axis=1)
y = df[['Premium Amount']]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=100, random_state=42)
}

# Train & evaluate models
results = {}

for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train.values.ravel()) 
    y_pred = model.predict(X_test)

    # Calculate evaluation metrics
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    results[name] = {"MAE": mae, "RMSE": rmse, "R² Score": r2}
    print(f"{name} - MAE: {mae:.2f}, RMSE: {rmse:.2f}, R² Score: {r2:.2f}\n")


# Convert results to DataFrame and display
results_df = pd.DataFrame(results).T
print("\nModel Performance Summary:")
print(results_df)

results_df.plot(kind='bar', figsize=(10, 6))
plt.title("Model Performance Comparison")
plt.xlabel("Model")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.legend(["MAE", "RMSE", "R² Score"])
plt.show()
```

```python
import joblib  


best_model = XGBRegressor(n_estimators=100, random_state=42)
best_model.fit(X_train, y_train.values.ravel()) 


joblib.dump(best_model, "xgboost_model.pkl")
print(" XGBoost model saved successfully as 'xgboost_model.pkl'")


loaded_model = joblib.load("xgboost_model.pkl")
prediction = loaded_model.predict(X_test)  
```
